```python
#Voters eligibilty System
print(""" Please enter your personal information to check your eligibility to vote.
Ensure to fill in the right and required information.""")
full_name = input("Enter your full name: ")
age = int(input("Enter your age: "))
gender = input("Enter your gender (M or F): ")
nationality = input("Enter your nationality: ")
residency = input("Enter your place of residence: ")
registration_status = input("Enter your registration status: ")
legal_restriction = input("Enter your legal restriction if any: ")

if age >= 18 and nationality.capitalize() == "Nigerian" and residency.capitalize() == "Nigeria" and  registration_status.capitalize() == "Registered" and legal_restriction.capitalize() =="None":
  print(f"{full_name} is eligible to vote.")
else:
  print(f"{full_name} is not eligible to vote.")
```
